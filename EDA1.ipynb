```python
df_country=pd.read_excel('/content/Country-Code.xlsx')
df_country.head()
```

```python
qw=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
qw.columns
```

```python
qw.Locality.value_counts()
```

```python
county_name=qw.Country.value_counts().index
```

```python
county_value=qw.Country.value_counts().values
```

```python
#pie chart top 5 countries
plt.pie(county_value[:3],labels=county_name[:3],autopct="%1.2f%%")
```

```python
#oberservation of zomato maximum records or transcation are from india/usa/brazil
```

```python

rating=qw.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
#when rating b/w 4.8-4.9 --> excellent
#when rating b/w 4.0-3.5 --> very good
#when rating b/w 3.5-2.5 --> good
#when rating below 2.5 --> bad
rating.head()
```

```python
plt.rcParams['figure.figsize']=(15,8)
sns.barplot(x='Aggregate rating', y='Rating Count', data=rating)
```

```python
sns.barplot(x='Aggregate rating', y='Rating Count',hue='Rating color',data=rating,palette=['grey','red','orange','yellow','green','green'])
```

```python
#OBSERVATION 1 NOT RATED MAXIMAUM ABOVE 2000
#MAXIMU RATING B/W 2.5-3.5
```

```python
#COUNT PLOT
sns.countplot(x='Rating color', data=rating, palette=['grey','red','orange','yellow','green','green']) # Removed y='Rating color Count'
```
